```python
import datetime as datetime
```

```python
df1['Purchase_Date'] = pd.to_datetime(df1['Purchase_Date'])
```

```python
df1.info()
```

```python
import numpy as np
```

```python
max_purday = np.max(df1['Purchase_Date'])
```

```python
df1_recency = df1
```

```python
df1_recency
```

```python
df1_recency = df1_recency.groupby("CustomerID")['Purchase_Date'].max().reset_index()
```

```python
df1_recency
```

```python
df1_recency["date_diff"] = df1_recency["Purchase_Date"].apply(lambda x: max_purday - x)
```

```python
df1_recency
```

```python
max_purday
```

```python
df1_recency = df1_recency[df1_recency["CustomerID"]!=0]
```

```python
df1_recency
```

```python
#quantile date_diff
q1 = df1_recency["date_diff"].quantile(0.25)
q2 = df1_recency["date_diff"].quantile(0.5)
q3 = df1_recency["date_diff"].quantile(0.75)
```

```python
print(q1,q2,q3)
```

```python
q1 = q1.days
q2 = q2.days
q3 = q3.days
```

```python
df1_recency.info()
```

```python
df1_recency['date_diff'] = df1_recency["date_diff"].dt.days
```

```python
df1_recency
```

```python
df1_recency["R"] = df1_recency["date_diff"].apply(lambda x: 4 if x <q1 else(3 if (x>=q1) and (x<q2)else( 2 if(x>=q2) and (x<q3) else(1))))
```

```python
df1_recency
```

```python
df1_recency["R"].unique()
```

```python
df1
```

```python
df1_frequency = df1.groupby("CustomerID")["CustomerID"].count().reset_index(name="frequency")
```

```python
df1_frequency
```

```python
df1_frequency = df1_frequency[df1_frequency["CustomerID"]!=0]
```

```python
df1_frequency
```

```python
df1_frequency.groupby("frequency")['CustomerID'].count().reset_index()
```

```python
q1 = df1_frequency['frequency'].quantile(0.25)
q2 = df1_frequency['frequency'].quantile(0.5)
q3 = df1_frequency['frequency'].quantile(0.75)
```

```python
print(q1,q2,q3)
```

As you can see that the q1, q2, and q3 have the same value so we can you quantile to do customer segmentation. Instead, we will assign q1, q2, q3 based on the enterprise and the industry

```python
q1 = 2
q2 = 3
q3 = 4
```

```python
df1_frequency
```

```python
df1_frequency["F"] = df1_frequency["frequency"].apply(lambda x: 4 if x>q3 else(3 if(x<=q3) and (x>q2) else(2 if (x<=q2) and (x>q1) else(1))))
```

```python
df1_frequency
```

```python
df1_frequency.groupby("F")["frequency"].sum().reset_index(name="frequency")
```

```python
df1
```

```python
df1_monetary = df1.groupby("CustomerID")["GMV"].sum().reset_index()
```

```python
df1_monetary
```

```python
df1_monetary = df1_monetary[df1_monetary['CustomerID']!=0]
```

```python
df1_monetary
```

```python
q1= df1_monetary["GMV"].quantile(0.25)
q2= df1_monetary["GMV"].quantile(0.5)
q3= df1_monetary["GMV"].quantile(0.75)
```

```python
print(q1,q2,q3)
```

```python
df1_monetary["M"] = df1_monetary["GMV"].apply(lambda x: 4 if x>q3 else(3 if (x<=q3) and (x>q2) else(2 if(x<=q2) and (x>q1) else(1))))
```

```python
df1_monetary
```

```python
df1_monetary["M"].unique()
```

```python
df1_recency
```

```python
df1_frequency
```

```python
df1_monetary
```

```python
df2 = pd.merge(pd.merge(df1_recency,df1_frequency,how='inner',on="CustomerID"),df1_monetary,how="inner",on="CustomerID")
```

```python
df2
```

```python
df2.info()
```

```python
df2["R"] = df2["R"].astype("str")
df2["F"] = df2["F"].astype("str")
df2["M"] = df2["M"].astype("str")
```

```python
df2.info()
```

```python
df2["RFM"] = df2["R"] + df2["F"] + df2["M"]
```

```python
df2
```

```python
df2.groupby("RFM")['RFM'].count().reset_index(name="number").sort_values(by=["number"],ascending=False)
```

```python
df2.to_csv("OutputRFM.csv",index=False)
```
